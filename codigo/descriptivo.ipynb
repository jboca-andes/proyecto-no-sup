```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Carga de datos
premier = pd.read_excel('Premier League 2011-12 Match by Match.xls')
```

```python
#Forma de los datos
print(premier.shape)
premier.head()
```

```python
#EstadÃ­sticas descriptivas
stats = premier.describe()
stats
```

```python
#Cantidad de jugadores
premier['Player ID'].unique().shape
```

```python
#Cantidad de equipos
premier['Team'].unique().shape
```

```python
#Variables sin variabilidad
sin_varianza = stats.loc['std',:] == 0
stats.loc[:,sin_varianza]
```

```python
#Ejemplo multicolinealidad perfecta 
#Total Successful Passes All = Successful Passes Defensive third + Successful Passes Middle third + Successful Passes Final third

premier[['Total Successful Passes All','Successful Passes Defensive third','Successful Passes Middle third', 'Successful Passes Final third']]
```

```python
# Ejemplo variable relevante para Delanteros

sns.set(rc={'figure.figsize':(8,4.5)})
ax = sns.violinplot(x='Position Id', y='Shots On Target inc goals', hue='Venue', data=premier, split=True, inner='quartile', palette='Set2')
ax.set_xticklabels(['Arquero','Defensa','Mediocampo','Delantero'])
plt.show()
```

```python
# Ejemplo variable relevante para Mediocampo

sns.set(rc={'figure.figsize':(8,4.5)})
ax = sns.violinplot(x='Position Id', y='Successful Passes Final third', hue='Venue', data=premier, split=True, inner='quartile', palette='Set2')
ax.set_xticklabels(['Arquero','Defensa','Mediocampo','Delantero'])
plt.show()
```

```python
# Ejemplo variable relevante para Defensa

sns.set(rc={'figure.figsize':(8,4.5)})
ax = sns.violinplot(x='Position Id', y='Total Clearances', hue='Venue', data=premier, split=True, inner='quartile', palette='Set2')
ax.set_xticklabels(['Arquero','Defensa','Mediocampo','Delantero'])
plt.show()
```

```python
# Ejemplo variable relevante para Arquero

sns.set(rc={'figure.figsize':(8,4.5)})
ax = sns.violinplot(x='Position Id', y='Catches', hue='Venue', data=premier, split=True, inner='quartile', palette='Set2')
ax.set_xticklabels(['Arquero','Defensa','Mediocampo','Delantero'])
plt.show()
```

Correlaciones

```python
# Se eliminan variables que no cambian
sin_varianza[sin_varianza].index
datos_num = premier.drop(sin_varianza[sin_varianza].index, axis=1)

#Se seleccionan las variables numÃ©ricas
datos_num = datos_num.iloc[:,10:]
```

```python
#CÃ¡lculo matriz de correlaciones
corr = datos_num.corr()

sns.set(rc={'figure.figsize':(32,18)})
sns.heatmap(corr)
plt.show()
```

```python
#AgrupaciÃ³n de datos por tipo de jugador calculando promedios
agrupados = premier.groupby('Player ID').mean()

#Se seleccionan delanteros
delanteros = agrupados.loc[agrupados['Position Id']==4, :]
delanteros
```

```python
#Ejemplo diagramas de dispersion para delanteros.

sns.set(rc={'figure.figsize':(16,9)})

sns.pairplot(delanteros.loc[:,['Goals', 'Total Successful Passes All', 'Assists' , 'Time Played' ]], y_vars=['Goals'])
plt.show()
```
